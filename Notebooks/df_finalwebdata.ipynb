```python
def error_steps(df):
    list_steps = []

    for i in range(len(df["process_step"]) - 1):
        
        if df["process_step"].iloc[i+1] < df["process_step"].iloc[i]:
            list_steps.append(df["process_step"].iloc[i])
    return list_steps
```
